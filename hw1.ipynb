## Problem 1

Make a tuple containing natural numbers, the square of which is a multiple of 3, 4, but not a multiple of 8 and not exceeding 12345.

````python
# Your solution here
lst = []
i = 6
while i ** 2 <= 12345:
    x = i ** 2
    if x % 12 == 0 and x % 8 != 0:
        lst.append(i)
    i += 1
tpl = tuple(lst)
print(tpl)
````

## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
# Your solution here
def rotate(matrix, direction):
    if not matrix or not matrix[0]:   #not empty input check
        print("matrix incorrect")
        return []

    if direction == 'right':
        return [list(row) for row in zip(*matrix[::-1])]
    elif direction == 'left':
        return [list(row) for row in zip(*matrix)][::-1]
    else:
        print("direction incorrect")
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(*rotate(matrix, 'left'))
print(*rotate(matrix, 'right'))
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
# Your solution here
class Book:
    def __init__(self, title, author, year, isbn, number):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.number = number
        print(f"Book {title} created!")

    def get_title(self):
        return self.title

    def get_author(self):
        return self.author

    def get_year(self):
        return self.year

    def get_isbn(self):
        return self.isbn

    def get_number(self):
        return self.number

    def change_number(self, status):
        if status == "borrow":
            if self.number > 0:
                self.number -= 1
                return True
            return False

        elif status == "return":
            self.number += 1
            return True

        else:
            raise ValueError("Error: Incorrect status!")

class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.borrowed_books = []
        print(f"User {name} created!")

    def change_info(self, new_name = None, new_card_number = None):
        if new_name:
            self.name = new_name

        if new_card_number:
            self.card_number = new_card_number

    def borrow_book(self, book):
        if book.get_number() > 0:
            self.borrowed_books.append(book)
            book.change_number("borrow")
            return True
        return False

    def return_book(self, book):
        self.borrowed_books.remove(book)
        book.change_number("return")
        return True

class Library:
    def __init__(self):
        self.users = []
        self.books = []
        self.history = []
        print("Library created!")

    def add_user(self, user):
        for lib_user in self.users:
            if lib_user.card_number == user.card_number:
                print("User already exists!")
                return False
        self.users.append(user)
        return True

    def remove_user(self, user):
        if user not in self.users:
            print("User not found!")
            return False
        if len(user.borrowed_books) > 0:
            print("The user did not return the books!")
            return False
        self.users.remove(user)
        return True

    def add_book(self, book):
        for lib_book in self.books:
            if lib_book.isbn == book.isbn:
                print("Book already exists!")
                return False
        self.books.append(book)
        return True

    def remove_book(self, book):
        if book not in self.books:
            print("Book not found!")
            return False
        self.books.remove(book)
        return True

    def borrow_book(self, user, book):
        if user.borrow_book(book):
            self.history.append(f"{user.name} borrowed {book.title}")
            return True
        return False

    def return_book(self, user, book):
        if book in user.borrowed_books:
            if user.return_book(book):
                self.history.append(f"{user.name} return {book.title}")
                return True
        print("The user did not borrow this book!")
        return False

    def find_book(self, title = None, author = None, year = None, isbn = None):
        result = self.books
        if title:
            result = [book for book in result if title.lower() in book.get_title().lower()]
        if author:
            result = [book for book in result if author.lower() in book.get_author().lower()]
        if year:
            result = [book for book in result if year == book.get_year()]
        if isbn:
            result = [book for book in result if isbn == book.get_isbn()]

        print(f"Total books found: {len(result)}")
        for book in result:
            print (f"Title: {book.title}, Author: {book.author}, Year: {book.year}, ISBN: {book.isbn}, Copies: {book.number}")

    def find_user(self, name = None, card_number = None):
        result = self.users
        if name:
            result = [user for user in result if name.lower() in user.name.lower()]
        if card_number:
            result = [user for user in result if card_number == user.card_number]

        print(f"Total users found: {len(result)}")
        for user in result:
            print(f"Name: {user.name}, Card number: {user.card_number}, Borrowed books: {[book.title for book in user.borrowed_books]}")

    def display_history(self):
        for record in self.history:
            print(record)

system = Library()
book1 = Book("Pilgrimage to  Earth", "Robert Sheckley", 1956, "9780899683621", 1)
system.add_book(book1)

book2 = Book("Ваш выход, или Шутов хоронят за оградой", "Генри Лайон Олди", 2001, "5-04-009829-4", 3)
system.add_book(book2)



user1 = User("Kerkilas Andros", 123456)
system.add_user(user1)

user2 = User("Dmitry Gromov", 654321)
system.add_user(user2)



system.borrow_book(user1, book1)

system.find_book(author = "E.B. Vinberg")
system.find_user(card_number = 123456)


system.return_book(user1, book1)

system.display_history()
````

## Problem 6*

Let 
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python
# Your solution here
ans1 = 0
ans2 = 0
for i in range(1, 10001):
    ans1 += 1 / i ** 2
    ans2 += 1 / (10001 - i) ** 2
print("A = ", ans1)
print("B = ", ans2)
if ans1 > ans2:
    print("A > B")
    print("difference =  ", ans1 - ans2)
else:
    print("A < B")
    print("difference = ", ans2 - ans1)
````
